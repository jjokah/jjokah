## Objects and variables

Everything in Python is an object, a piece of data stored in memory. A variable is just a name you attach to that object so you can refer to it later.

Think of it like a flashcard: you write a label on the front and the value on the back. When Python needs to recall it, it looks up the label.

````python
name = "JohnJohnson"
score = 95
is_student = True
````

A few rules:
- Variable names are **case-sensitive**: `username` and `Username` are different things
- Names can't start with a number

## Checking types

Every object in Python has a type. You can check it with the built-in `type()` function.

````python
print(type("JohnJohnson"))   # <class 'str'>
print(type(95))              # <class 'int'>
print(type(3.14))            # <class 'float'>
print(type(True))            # <class 'bool'>
````

This matters more than it sounds. A lot of bugs come from accidentally mixing types, such as treating a number as a string, or a string as a list. Getting in the habit of checking types early saves frustration later.

## Bugs

A bug is anything that causes your code to behave differently than you intended. There are two main kinds:

**Syntax errors**: Python can't even parse what you wrote. These are usually obvious: a missing colon, mismatched quotes, wrong indentation.

````python
# Missing colon -- this will throw a SyntaxError:
# if x == 5
#     print("five")
````

**Logic errors**: Python runs fine but produces the wrong output. These are sneakier because Python won't tell you anything is wrong.

````python
# This runs, but it's wrong -- x is 10, not 5, so nothing prints
x = 10
if x == 5:
    print("five")
````

Debugging is a skill, not a sign of failure. Every programmer spends a significant portion of their time tracking down bugs, including experienced ones. The goal isn't to avoid bugs entirely; it's to get better at finding them quickly.

When something breaks, read the error message carefully. Python's error messages are actually pretty descriptive. The line number it points to is where Python noticed the problem, which is usually close to (though not always exactly at) where the mistake is.

# Data and Data Structures

## Strings

A string is text: any sequence of characters wrapped in quotes. Single or double, Python accepts both.

````python
message = "Hello, world"
title = "Monty Python and the Holy Grail"
````

Strings are **immutable**: once created, you can't change the value in memory. What you can do is create a new string based on the old one and reassign the variable.

````python
status = "pending"
status = status.upper()
print(status)  # PENDING
````

### Useful string methods

````python
name = "  elon musk  "

print(name.strip())                   # "elon musk" (removes leading/trailing whitespace)
print(name.lower())                   # "elon musk"
print(name.upper())                   # "ELON MUSK"
print(name.capitalize())              # "Elon musk"

print("elon" in name)                 # True (membership test)
print(name.replace("elon", "jane"))   # "  jane musk  "
print(name.split(" "))                # ['', '', 'elon', 'musk', '', '']
````

Case matters when searching. `"Elon"` and `"elon"` are not the same string. If you're checking whether a name or word appears in some text, normalize the case first with `.lower()`, otherwise you'll miss matches just because of capitalization.

## Numbers

Python has two main numeric types:

- **Integers** (`int`): whole numbers: `42`, `-7`, `0`
- **Floats** (`float`): decimals: `3.14`, `-0.5`, `1.0`

````python
year = 2024
price = 9.99

print(10 + 3)    # 13    (addition)
print(10 - 3)    # 7     (subtraction)
print(10 * 3)    # 30    (multiplication)
print(10 ** 3)   # 1000  (exponentiation: 10 to the power of 3)
print(10 / 3)    # 3.3333... (division, always returns float)
print(10 // 3)   # 3     (floor division, integer result)
print(10 % 3)    # 1     (modulo, remainder)
````

The modulo operator (`%`) is more useful than it looks. It's how you check if a number is even/odd, or whether something divides evenly.

## Data Structures

Once you have data, you need ways to hold multiple pieces of it together. Python gives you four core structures.

### Lists `[]`

A list is an ordered, mutable collection. You can change it after creating it: add items, remove items, update items. Lists are **zero-indexed**: the first item is at position 0.

````python
scores = [82, 91, 78, 95]

print(scores[0])     # 82
print(scores[1])     # 91
print(scores[-1])    # 95 (negative index counts from the end)
print(scores[1:3])   # [91, 78] (slicing: from index 1 up to, not including, 3)

scores.append(88)
print(scores)  # [82, 91, 78, 95, 88]
````

### Tuples `()`

Tuples look like lists but they're **immutable**: you can't modify them after creation. Use tuples when the data shouldn't change: coordinates, fixed configuration pairs, function return values.

````python
coordinate = (40.7128, -74.0060)
print(coordinate[0])   # 40.7128  (latitude)
print(coordinate[1])   # -74.0060 (longitude)
````

If you try to modify a tuple, Python throws an `AttributeError`. That's intentional: it's a signal that this data is meant to stay fixed.

### Sets `{}`

A set is a mutable collection with **no duplicate values**. Useful for deduplication and membership testing.

````python
genres = {"fiction", "mystery", "fiction", "sci-fi"}
print(genres)  # {'fiction', 'mystery', 'sci-fi'} (duplicates removed)
````

Sets are unordered, so you can't index into them. They're best when you need to know if something is present or when you want to work with unique values only: a list of tags, categories, names without repeats.

### Dictionaries `{}`

Dictionaries store **key-value pairs**. Instead of using a numeric index to look up data, you use a meaningful key.

````python
book = {
    "title": "The Hitchhiker's Guide to the Galaxy",
    "author": "Douglas Adams",
    "year": 1979,
    "available": True
}

print(book["title"])    # The Hitchhiker's Guide to the Galaxy
print(book["author"])   # Douglas Adams

book["rating"] = 4.8
book["available"] = False
print(book)
````

Compared to a list, a dictionary is self-documenting. `book["author"]` tells you exactly what you're looking at; `book[2]` tells you nothing. For anything with named fields like a contact, a product, or a configuration, dictionaries are almost always the better choice.

---

## When to use which structure

| Structure | Use when... |
|-----------|-------------|
| `list` | You have an ordered collection you'll modify: adding, removing, iterating |
| `tuple` | The data is fixed and shouldn't change: config pairs, coordinates, return values |
| `set` | You need uniqueness: deduplication, fast membership checks |
| `dict` | You have named fields: records, config objects, anything structured |

There's no single right answer for every situation, but this table covers the vast majority of decisions you'll face.

# Control Flow: Loops and Conditionals

Control flow is how you tell your program to make decisions and repeat actions. Without it, your code runs top to bottom once and stops. With it, you can process a thousand records with a few lines, or react differently based on what the data says.

---

## For loops

A `for` loop iterates over a sequence (a list, a string, a range of numbers) and runs a block of code once for each item.

Before writing the code, say it in plain language: *"For each name in this list, print it."*

````python
names = ["Alice", "Bob", "Charlie", "Diana"]

for name in names:
    print(name)
````

The variable name after `for` (here `name`) is just a label: it takes on each value in the sequence one at a time. Name it something meaningful; it makes the code easier to read.

You can do more than just print:

````python
names = ["Alice", "Bob", "Charlie"]
greetings = []

for name in names:
    greetings.append(f"Hello, {name}!")

print(greetings)
````

### Enumerate

When you need both the index and the value, use `enumerate()`. It gives you a counter for free without having to manually track it.

````python
for i, name in enumerate(names):
    print(i, name)
````

### List comprehension

List comprehension is a compact way to build a list from a loop in a single line. Both versions below produce the same result. Comprehensions are more Pythonic and concise, but for beginners, the verbose version is easier to read and debug. Use comprehensions once you're comfortable with what they're doing under the hood.

````python
# Verbose version
greetings = []
for name in names:
    greetings.append(f"Hello, {name}!")

# List comprehension version
greetings = [f"Hello, {name}!" for name in names]
print(greetings)
````

---

## While loops

A `while` loop keeps running as long as a condition is true. While loops are best when you don't know in advance how many iterations you need, such as waiting for a connection, retrying a request, or reading input until a condition is met. Be careful: if the condition never becomes false, you'll have an infinite loop. Always make sure something inside the loop moves it toward termination.

````python
attempts = 0

while attempts < 3:
    print(f"Attempt {attempts + 1}")
    attempts += 1
````

---

## Conditionals

Conditionals let your code branch based on what's true. The structure is `if`, `elif` (else if), and `else`.

### if / else

````python
temperature = 38

if temperature > 37.5:
    print("Fever detected")
else:
    print("Temperature is normal")
````

### if / elif / else

When you have more than two cases, chain them with `elif`. Python checks each condition top to bottom and executes the first one that's true, then skips the rest. Order matters: put the most specific or most likely conditions first.

````python
age = 20

if age >= 65:
    print("Senior discount applies")
elif age >= 18:
    print("Adult price")
elif age >= 12:
    print("Teen price")
else:
    print("Child price")
````

### Comparison operators

| Operator | Meaning |
|----------|---------|
| `x == y` | equal to |
| `x != y` | not equal to |
| `x > y`  | greater than |
| `x < y`  | less than |
| `x >= y` | greater than or equal to |
| `x <= y` | less than or equal to |

### in / not in

`in` checks whether a value exists inside a sequence, and works with strings, lists, and sets. `not in` is the inverse.

````python
allowed_colors = ["red", "green", "blue"]
chosen = "green"

if chosen in allowed_colors:
    print("Valid color selection")

sentence = "The quick brown fox"

if "fox" in sentence:
    print("Found it")

if "cat" not in sentence:
    print("No cat in this sentence")
````

---

## Combining loops and conditionals

The real power kicks in when you combine the two. This pattern of looping over a collection and conditionally acting on each item is everywhere in real code.

````python
books = [
    {"title": "Dune", "author": "Frank Herbert", "pages": 412, "read": True},
    {"title": "1984", "author": "George Orwell", "pages": 328, "read": False},
    {"title": "Sapiens", "author": "Yuval Noah Harari", "pages": 443, "read": True},
    {"title": "The Alchemist", "author": "Paulo Coelho", "pages": 208, "read": False},
]

unread = []

for book in books:
    if not book["read"]:
        unread.append(book["title"])

print(f"You have {len(unread)} unread book(s):")
for title in unread:
    print(f" - {title}")
````

This exact pattern of looping over a collection, filtering by condition, and collecting results is something you'll write in some variation constantly, whether you're processing a list of items, records from a file, responses from an API, or rows in a spreadsheet.

# Functions, Classes, and Libraries

Up until now, everything we've written has been procedural: code that runs top to bottom, once. That's fine for small scripts. But as soon as you're doing the same thing in multiple places, or your code starts getting long enough that you have to scroll to remember what's at the top, it's time to organize.

---

## Functions

A function is a reusable block of code with a name. You define it once, and call it as many times as you need.

The anatomy of a function:
- `def`: the keyword that starts a function definition
- the function name
- `(parameters)`: the inputs the function expects
- the indented block: the function body
- `return`: what the function sends back when it's done

````python
def greet_user(username):
    message = f"Welcome back, {username}."
    return message

result = greet_user("jjokah")
print(result)   # Welcome back, jjokah.
````

### Multiple parameters

Arguments are matched by position. The first value goes to the first parameter, second to second, and so on.

````python
def full_name(first, last):
    return f"{first} {last}"

print(full_name("JohnJohnson", "Okah"))  # JohnJohnson Okah
````

### Keyword arguments

You can also pass arguments by name, which lets you specify them in any order:

````python
print(full_name(last="Okah", first="JohnJohnson"))  # JohnJohnson Okah
````

### Default parameter values

If a parameter has a sensible default, you can bake it in. Parameters with defaults must come *after* parameters without defaults.

````python
def make_greeting(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(make_greeting("Alice"))                  # Hello, Alice!
print(make_greeting("Alice", "Good morning"))  # Good morning, Alice!
````

### Arbitrary arguments (`*args`)

When you don't know how many arguments will be passed, use `*args`. Python collects everything into a tuple.

````python
def list_items(*items):
    for item in items:
        print(f"- {item}")

list_items("apples", "bread", "milk", "eggs")
````

### Docstrings

A docstring is a short description placed right after the `def` line, wrapped in triple quotes. It's documentation for anyone (including future you) who uses the function. Docstrings aren't enforced by Python, but writing them is a habit worth building. Code without documentation is a liability.

````python
def mean(values):
    """Returns the arithmetic mean of a list of numbers."""
    return sum(values) / len(values)

print(mean.__doc__)
print(mean([10, 20, 30]))  # 20.0
````

## Classes

A class is a blueprint for creating objects that bundle data and behavior together. Where a dictionary holds data, a class holds data *and* functions that operate on that data.

A few things to unpack:
- `__init__` is the **initializer**: it runs automatically when you create an instance. This is where you set the object's initial state.
- `self` is a reference to the instance itself. Every method in a class takes `self` as its first parameter. When you call a method, Python passes the instance automatically.
- Attributes like `self.title` are data attached to the instance.

````python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.notes = []

    def add_note(self, note):
        self.notes.append(note)

    def summary(self):
        note_count = len(self.notes)
        return f'"{self.title}" by {self.author}, {note_count} note(s)'


book = Book("Dune", "Frank Herbert")
book.add_note("Ecology as a metaphor for power.")
book.add_note("Paul's prescience feels like a burden, not a gift.")

print(book.summary())
print(book.title)
print(book.notes)
print(vars(book))
````

### When to use a class vs. a dict vs. a function

- Use a **function** when you have a reusable operation that takes inputs and returns an output
- Use a **dict** when you have a simple record with named fields and no associated behavior
- Use a **class** when you have structured data *and* behavior that belongs with it, when it makes sense to say "this object *does* things"

A book's title and author stored once? A dict is fine. A book object that tracks notes, marks chapters read, and exports a summary? That's a class.

## Reading and writing text files

Python's built-in `open()` function gives you access to files. The right way to use it is with the `with` statement, which is called a **context manager**.

The `with` block automatically closes the file when it exits, even if something goes wrong. Without it, you'd have to call `f.close()` manually, and if your code throws an error before that line, the file stays open. The context manager pattern is cleaner and safer.

The second argument to `open()` is the **mode**:
- `"r"`: read (default)
- `"w"`: write (creates the file or overwrites it if it exists)
- `"a"`: append (adds to the end without overwriting)

### Reading files

`.splitlines()` is better than `.split("\\n")` for most cases: it handles different line endings (Windows uses `\\r\\n`, Unix uses `\\n`) without you having to think about it.

````python
# Read the entire file as one string
with open("notes.txt", "r") as f:
    content = f.read()

# Read line by line into a list
with open("notes.txt", "r") as f:
    lines = f.read().splitlines()

print(lines[0])   # first line
print(lines[-1])  # last line
````

### Writing files

If the file already exists, `"w"` mode overwrites it entirely. Use `"a"` if you want to add to an existing file without destroying what's there.

````python
summary = "Report generated on 2026-03-19. 42 items reviewed."

with open("report.txt", "w") as f:
    f.write(summary)
````

## JSON

JSON (JavaScript Object Notation) is the standard format for structured, hierarchical data. If you've ever worked with an API, you've seen it. It maps almost directly onto Python's dictionaries and lists.

```json
{
  "title": "Dune",
  "author": "Frank Herbert",
  "year": 1965,
  "tags": ["sci-fi", "classic"]
}
```

Python's built-in `json` module handles it.

### Writing JSON to a file

`indent=4` makes the output human-readable. Without it, everything gets written on one line.

````python
import json

book = {
    "title": "Dune",
    "author": "Frank Herbert",
    "year": 1965,
    "tags": ["sci-fi", "classic"]
}

with open("book.json", "w") as f:
    json.dump(book, f, indent=4)
````

### Reading JSON from a file

Once loaded, `data` is a regular Python dictionary, so you work with it exactly the same way you'd work with any dict.

````python
with open("book.json", "r") as f:
    data = json.load(f)

print(data["title"])    # Dune
print(data["tags"])     # ['sci-fi', 'classic']
````

### JSON vs CSV

CSV is fine for flat tabular data (rows and columns, nothing nested). JSON handles nested and hierarchical data (lists inside dictionaries, dictionaries inside lists), which is most of what you encounter in real systems. If you're storing anything more complex than a spreadsheet row, JSON is the better choice.

## Working with multiple files

When you have a directory full of files to process, you have two tools: `glob` for simple pattern matching, and `os.walk` for deep nested directories.

### glob

`glob` finds files by pattern. It's in the standard library, so no install needed. The `*` is a wildcard: it matches any sequence of characters. `glob.glob()` returns a list of matching file paths as strings.

````python
import glob

# All files in a directory
all_files = glob.glob("data/*")

# Only .txt files
txt_files = glob.glob("data/*.txt")

# .txt files one directory deep
nested = glob.glob("data/*/*.txt")

print(txt_files)
````

````python
for filepath in glob.glob("books/*.txt"):
    with open(filepath, "r") as f:
        content = f.read()
    print(f"--- {filepath} ---")
    print(content[:200])   # first 200 characters
````

### os.walk

When your files are buried in multiple levels of subdirectories, `glob` with fixed depth patterns gets unwieldy. `os.walk` traverses an entire directory tree regardless of depth.

`os.walk` gives you three things at each level:
- `root`: the current directory path
- `subdirs`: list of subdirectory names inside it
- `files`: list of filenames inside it

Use `os.path.join(root, filename)` to build the full path safely, as it handles path separators correctly across operating systems.

````python
import os

for root, subdirs, files in os.walk("library/"):
    for filename in files:
        filepath = os.path.join(root, filename)
        print(filepath)
````

````python
# Filtering to specific file types
json_files = []

for root, subdirs, files in os.walk("library/"):
    for filename in files:
        if filename.endswith(".json"):
            json_files.append(os.path.join(root, filename))

print(f"Found {len(json_files)} JSON files")
````

### Putting it together

Here's a pattern you'll use constantly: process every file in a directory tree and collect results.

The `_` in `for root, _, files` is a convention for "I don't need this value": it still gets assigned, but the underscore signals to the reader that it's intentionally unused.

````python
import os
import json

all_books = []

for root, _, files in os.walk("library/"):
    for filename in files:
        if filename.endswith(".json"):
            filepath = os.path.join(root, filename)
            with open(filepath, "r") as f:
                data = json.load(f)
            all_books.append(data)

print(f"Loaded {len(all_books)} book records")
````

## How HTML works (the minimum you need)

HTML is a markup language that structures web pages using **tags**. Tags come in pairs, an opening tag and a closing tag, with content between them.

```html
<p>This is a paragraph.</p>
<strong>This is bold text.</strong>
```

Tags can be nested:

```html
<div>
    <p>A paragraph inside a div.</p>
</div>
```

Tags can also have **attributes**, which are metadata that give them additional properties. The two most common ones you'll encounter when scraping are `class` and `id`.

```html
<div class="article-body">
    <p id="intro">First paragraph.</p>
</div>
```

`class` groups elements together; `id` identifies one specific element. Both are useful for targeting exactly what you want when extracting data.

You don't need to write HTML. You just need to read it well enough to find what you're after.

## Fetching a page with requests

`requests.get()` sends an HTTP GET request and returns a response object. The key pieces:
- `response.status_code`: the HTTP status code (200 = OK, 403 = forbidden, 404 = not found, 429 = rate limited)
- `response.content`: the raw bytes of the page
- `response.text`: the page content decoded as a string

Always check the status code before trying to use the content. If you get anything other than 200, the data you have may be an error page, not the page you wanted.

````python
import requests

url = "https://en.wikipedia.org/wiki/List_of_French_monarchs"
response = requests.get(url)

print(response.status_code)   # 200 means success
print(type(response.content)) # <class 'bytes'>
````

## Parsing HTML with BeautifulSoup

BeautifulSoup takes raw HTML and turns it into a navigable object called a **soup**.

- `.find()` returns one element (or `None` if not found)
- `.find_all()` always returns a list, even if there's only one match
- `.text` strips all the HTML tags and returns the plain text content of the element

````python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, "html.parser")

# First occurrence of a tag
first_para = soup.find("p")

# All occurrences of a tag
all_paras = soup.find_all("p")

# Tag with a specific attribute
main = soup.find("div", {"id": "mw-content-text"})

for p in soup.find_all("p")[:3]:
    print(p.text)
````

### A complete example

Fetching and printing all paragraphs from a Wikipedia page:

````python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_French_monarchs"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    main_content = soup.find("div", {"id": "mw-content-text"})
    paragraphs = main_content.find_all("p")
    for p in paragraphs:
        text = p.text.strip()
        if text:   # skip empty paragraphs
            print(text)
else:
    print(f"Failed: {response.status_code}")
````

## Extracting tables

HTML tables follow a predictable structure:

```html
<table>
    <tr>
        <th>Name</th><th>Reign Start</th><th>Reign End</th>
    </tr>
    <tr>
        <td>Hugh Capet</td><td>987</td><td>996</td>
    </tr>
</table>
```

- `<table>`: the table container
- `<tr>`: a row
- `<th>`: a header cell
- `<td>`: a data cell

````python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_French_monarchs"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Wikipedia tables use the "wikitable" class
tables = soup.find_all("table", {"class": "wikitable"})
table = tables[0]

rows = table.find_all("tr")

for row in rows:
    headers = row.find_all("th")
    cells = row.find_all("td")

    if headers:
        print([h.text.strip() for h in headers])
    elif cells:
        print([c.text.strip() for c in cells])
````

### The shortcut: pandas.read_html

If you just need the table data and don't want to hand-parse the HTML, pandas can do it in one line. `pd.read_html()` finds all tables on the page and returns them as a list of DataFrames. It's fast and convenient, but gives you less control: citation numbers and footnote markers tend to end up in the data, and you can't target specific tables by CSS class the way BeautifulSoup can. Use whichever fits the task.

````python
import pandas as pd

tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_French_monarchs")
df = tables[0]
print(df.head())
````

## Being a responsible scraper

A few things worth knowing before you start scraping:

**Check the terms of service.** Some sites explicitly prohibit automated scraping. This matters less for public data and research, but it's worth being aware of.

**Respect rate limits.** If you're hitting a site in a loop, fetching hundreds of pages, add a small delay between requests. Hammering a server with requests is rude at best and gets your IP blocked at worst.

````python
import time

url_list = []  # your list of URLs here

for url in url_list:
    response = requests.get(url)
    # process response
    time.sleep(1)   # wait 1 second between requests
````

**Handle errors.** Pages go down. Network connections drop. Always check `response.status_code` and wrap requests in error handling for anything running in production or unattended.

````python
try:
    response = requests.get(url, timeout=10)
    response.raise_for_status()   # raises an exception for 4xx/5xx responses
except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
````

**Many modern sites use APIs.** If the site you're scraping has a public API, use it instead of scraping HTML. APIs give you structured data, are more stable (pages change, API contracts less so), and are typically the intended way to access the data.
